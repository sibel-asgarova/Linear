```python
import typing
from typing import List
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Load the dataframe
dataset=pd.read_csv("Salary_dataset.csv")
dataset.head(5)
```

```python
#Get information about the dataframe:
dataset.info()
```

```python
dataset.shape
```

```python
#Check the column names of the given dataframe
dataset.columns
```

```python
#Drop unnecessary columns:
dataset.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
dataset.head(5)
```

```python
#Divide dataset into X and y
X=dataset.drop("Salary",axis=1)
y=dataset.Salary
```

```python
#Implement Linear Regression model on the given dataframe
model=LinearRegression()
model.fit(X,y)
```

$$\bar{y} = b_1 \bar{x} + b_0$$

```python
print(model.coef_)      #b1    
print(model.intercept_) #b0
```

```python
b1=model.coef_
b0=model.intercept_
```

The equation with specific values for \( b1 \) and \( b0 \) is:

$$
\bar{y} = 9449.96232146 \cdot \bar{x} + 24848.203966523237
$$

```python
#Get predictions based on the Linear Regression model
y_pred=model.predict(X)
y_pred
```

```python
plt.scatter(X, y, color='blue', label='Actual Salary')
plt.plot(X, y_pred, color='red', label='Predicted Salary', linewidth=2)
plt.title('Actual vs Predicted Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
